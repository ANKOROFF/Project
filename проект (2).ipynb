## 2. Описание используемых методов НЛА
Для ускорения вычислений в ViT будут использованы следующие методы НЛА:
1. **Декомпозиция Холецкого** — метод для разложения положительно определенных матриц на произведение нижней и верхней треугольных матриц.
2. **Низкоранговые аппроксимации** — методы, такие как SVD, для приближения матриц с высокой размерностью более компактными матрицами.

4. **Быстрые преобразования Фурье (FFT)** — метод для ускорения вычислений в задачах обработки сигналов и изображений.
5. **Метод главных компонент (PCA)** — использование PCA для снижения размерности данных и ускорения расчетов.
6. **Метод QR-разложения** — для ускоренного решения линейных систем и вычислений с матрицами.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Setting device to CUDA if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Defining Vision Transformer class with a basic attention mechanism
class VisionTransformer(nn.Module):
    def __init__(self, dim, num_heads, num_classes, image_size, patch_size, num_layers):
        super(VisionTransformer, self).__init__()
        self.image_size = image_size
        self.patch_size = patch_size
        self.num_patches = (image_size // patch_size) ** 2
        self.dim = dim
        self.num_heads = num_heads
        self.num_classes = num_classes
        
        # Patch embedding layer
        self.patch_embed = nn.Conv2d(3, dim, kernel_size=patch_size, stride=patch_size)
        
        # Positional Encoding
        self.positional_encoding = nn.Parameter(torch.zeros(1, self.num_patches, dim))
        
        # Transformer Encoder
        self.encoder_layers = nn.ModuleList([
            nn.TransformerEncoderLayer(d_model=dim, nhead=num_heads) for _ in range(num_layers)
        ])
        
        # Classifier Head
        self.fc = nn.Linear(dim, num_classes)

    def forward(self, x):
        # Patch embedding
        x = self.patch_embed(x).flatten(2).transpose(1, 2)
        x = x + self.positional_encoding
        
        # Pass through Transformer Encoder layers
        for layer in self.encoder_layers:
            x = layer(x)
        
        # Pooling and Classification
        x = x.mean(dim=1)  # Global Average Pooling
        x = self.fc(x)
        return x


# Function to compute the self-attention scores
def compute_attention(query, key, value):
    # q, k, v are of shape (batch_size, num_patches, dim)
    attention_scores = torch.matmul(query, key.transpose(-2, -1)) / np.sqrt(query.size(-1))  # Scaled dot-product
    attention_probs = torch.softmax(attention_scores, dim=-1)
    attention_output = torch.matmul(attention_probs, value)
    return attention_output, attention_probs


# Cholesky Decomposition for Attention Mechanism
def cholesky_decompose_attention(query, key, value):
    # Perform Cholesky decomposition on the query and key matrices (reduce computational cost)
    Q = torch.matmul(query, query.transpose(-2, -1))  # Query matrix
    K = torch.matmul(key, key.transpose(-2, -1))  # Key matrix
    
    # Cholesky decomposition
    L_q = torch.linalg.cholesky(Q + 1e-5 * torch.eye(Q.size(-1)).to(device))  # Adding small identity matrix for numerical stability
    L_k = torch.linalg.cholesky(K + 1e-5 * torch.eye(K.size(-1)).to(device))  # Similarly for Key
    
    # Use decomposed matrices for attention calculation
    attention_scores = torch.matmul(L_q, L_k.transpose(-2, -1)) / np.sqrt(query.size(-1))
    attention_probs = torch.softmax(attention_scores, dim=-1)
    attention_output = torch.matmul(attention_probs, value)
    return attention_output, attention_probs


# Create an example input image
image_size = 224
patch_size = 16
dim = 256
num_heads = 8
num_layers = 12
num_classes = 1000

# Example input: Random image batch of size 8 (batch_size, channels, height, width)
x = torch.randn(8, 3, image_size, image_size).to(device)

# Create Vision Transformer instance
vit = VisionTransformer(dim=dim, num_heads=num_heads, num_classes=num_classes, image_size=image_size, patch_size=patch_size, num_layers=num_layers).to(device)

# Perform a forward pass
output = vit(x)
print(f"Output shape: {output.shape}")

# Visualize attention map (simplified example)
query = torch.randn(8, 196, dim).to(device)  # 196 tokens for 224x224 images with 16x16 patches
key = query  # In self-attention, query and key have the same shape
value = torch.randn(8, 196, dim).to(device)

# Apply Cholesky decomposition to attention mechanism
attention_output, attention_probs = cholesky_decompose_attention(query, key, value)
print(f"Attention output shape after Cholesky decomposition: {attention_output.shape}")

# Visualize the attention probabilities
plt.imshow(attention_probs[0].cpu().detach().numpy(), cmap='viridis')
plt.colorbar()
plt.title("Attention Map after Cholesky Decomposition")
plt.show()

```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

# Функция для применения SVD к матрице
def svd_approximation(matrix, rank):
    """
    Аппроксимация матрицы с использованием SVD для каждого батча.
    :param matrix: Исходная матрица.
    :param rank: Число сингулярных значений для аппроксимации.
    :return: Аппроксимированная матрица.
    """
    batch_size, seq_len, embed_dim = matrix.shape
    # Список для хранения аппроксимированных матриц
    approx_matrices = []

    for i in range(batch_size):
        # Извлекаем данные для текущего батча
        mat = matrix[i]  # Размерность (seq_len, embed_dim)
        
        # Выполняем SVD
        U, S, Vh = torch.linalg.svd(mat, full_matrices=False)
        
        # Сохраняем только первые "rank" сингулярных значений
        U_approx = U[:, :rank]
        S_approx = torch.diag(S[:rank])  # Диагональная матрица из первых "rank" сингулярных значений
        Vh_approx = Vh[:rank, :]
        
        # Формируем аппроксимированную матрицу
        approx_matrix = torch.matmul(U_approx, torch.matmul(S_approx, Vh_approx))
        approx_matrices.append(approx_matrix)

    # Конкатенируем результат по размерности батча
    return torch.stack(approx_matrices)

# Параметры
batch_size = 4
seq_len = 128
embed_dim = 256
rank = 50  # Используем более низкий ранг для аппроксимации

# Генерация случайных данных для запросов (Q), ключей (K) и значений (V)
Q = torch.randn(batch_size, seq_len, embed_dim)
K = torch.randn(batch_size, seq_len, embed_dim)
V = torch.randn(batch_size, seq_len, embed_dim)

# Применение SVD для аппроксимации матриц
Q_approx = svd_approximation(Q, rank)
K_approx = svd_approximation(K, rank)
V_approx = svd_approximation(V, rank)

# Рассчитываем внимание с использованием аппроксимированных матриц
attention_scores = torch.matmul(Q_approx, K_approx.transpose(-2, -1)) / np.sqrt(embed_dim)
attention_weights = F.softmax(attention_scores, dim=-1)

# Получаем результат внимания
output = torch.matmul(attention_weights, V_approx)

# Визуализируем карту внимания
plt.figure(figsize=(6, 6))
plt.imshow(attention_weights[0].detach().cpu().numpy(), cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Attention Map with SVD Approximation')
plt.xlabel('Keys')
plt.ylabel('Queries')
plt.show()
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

# Параметры
batch_size = 4
seq_len = 128
embed_dim = 256
rank = 50  # Используем более низкий ранг для аппроксимации
num_heads = 8  # Число голов в многоголовочном внимании

# Генерация случайных данных для запросов (Q), ключей (K) и значений (V)
Q = torch.randn(batch_size, seq_len, embed_dim)
K = torch.randn(batch_size, seq_len, embed_dim)
V = torch.randn(batch_size, seq_len, embed_dim)

# Применение SVD для аппроксимации матриц
def svd_approximation(matrix, rank):
    batch_size, seq_len, embed_dim = matrix.shape
    approx_matrices = []

    for i in range(batch_size):
        mat = matrix[i]
        U, S, Vh = torch.linalg.svd(mat, full_matrices=False)
        
        # Оставляем только первые "rank" сингулярных значений
        U_approx = U[:, :rank]
        S_approx = torch.diag(S[:rank])
        Vh_approx = Vh[:rank, :]
        
        # Формируем аппроксимированную матрицу
        approx_matrix = torch.matmul(U_approx, torch.matmul(S_approx, Vh_approx))
        approx_matrices.append(approx_matrix)

    return torch.stack(approx_matrices)

Q_approx = svd_approximation(Q, rank)
K_approx = svd_approximation(K, rank)
V_approx = svd_approximation(V, rank)

# Вычисление внимания
def compute_attention(Q, K, V, d_k):
    # Скаларное произведение запросов и ключей
    attention_scores = torch.matmul(Q, K.transpose(-2, -1)) / np.sqrt(d_k)
    
    # Применяем softmax для нормализации внимания
    attention_weights = F.softmax(attention_scores, dim=-1)
    
    # Вычисляем итоговое значение внимания
    output = torch.matmul(attention_weights, V)
    return output, attention_weights

# Многоголовочное внимание
def multi_head_attention(Q, K, V, num_heads, d_k):
    # Разделяем запросы, ключи и значения на несколько голов
    batch_size, seq_len, embed_dim = Q.shape
    head_dim = embed_dim // num_heads  # Размерность каждой головы
    
    # Изменяем форму данных для разделения на головы
    Q_ = Q.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    K_ = K.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    V_ = V.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    
    # Вычисляем внимание для каждой головы
    attention_outputs = []
    attention_weights_list = []
    
    for i in range(num_heads):
        output, attention_weights = compute_attention(Q_[:, i], K_[:, i], V_[:, i], d_k)
        attention_outputs.append(output)
        attention_weights_list.append(attention_weights)
    
    # Объединяем результаты всех голов
    attention_output = torch.cat(attention_outputs, dim=-1)  # Сшиваем по размерности "embed_dim"
    
    # Линейное преобразование (процесс после многоголовочного внимания)
    output = attention_output.view(batch_size, seq_len, embed_dim)
    
    return output, attention_weights_list

# Размерность ключей
d_k = Q.shape[-1]

# Вычисление многоголовочного внимания
output, attention_weights_list = multi_head_attention(Q_approx, K_approx, V_approx, num_heads, d_k)

# Визуализация карты внимания для первой головы
plt.figure(figsize=(6, 6))
plt.imshow(attention_weights_list[0][0].detach().cpu().numpy(), cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Attention Map for Head 1 (SVD Approximation)')
plt.xlabel('Keys')
plt.ylabel('Queries')
plt.show()

# Вывод итогового результата многоголовочного внимания
print("Output shape after multi-head attention:", output.shape)
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

# Параметры
batch_size = 4
seq_len = 128
embed_dim = 256
rank = 50  # Используем более низкий ранг для аппроксимации
num_heads = 8  # Число голов в многоголовочном внимании

# Генерация случайных данных для запросов (Q), ключей (K) и значений (V)
Q = torch.randn(batch_size, seq_len, embed_dim)
K = torch.randn(batch_size, seq_len, embed_dim)
V = torch.randn(batch_size, seq_len, embed_dim)

# Позиционные коды
def get_positional_encoding(seq_len, embed_dim):
    pos = torch.arange(0, seq_len).unsqueeze(1).float()  # (seq_len, 1)
    i = torch.arange(0, embed_dim, 2).float()  # (embed_dim // 2,)
    angles = pos / torch.pow(10000, (i / embed_dim))  # (seq_len, embed_dim // 2)

    positional_encoding = torch.zeros(seq_len, embed_dim)  # (seq_len, embed_dim)
    positional_encoding[:, 0::2] = torch.sin(angles)  # Синус для четных индексов
    positional_encoding[:, 1::2] = torch.cos(angles)  # Косинус для нечетных индексов

    return positional_encoding

# Создание позиционных кодов
positional_encoding = get_positional_encoding(seq_len, embed_dim)

# Применение SVD для аппроксимации матриц
def svd_approximation(matrix, rank):
    batch_size, seq_len, embed_dim = matrix.shape
    approx_matrices = []

    for i in range(batch_size):
        mat = matrix[i]
        U, S, Vh = torch.linalg.svd(mat, full_matrices=False)
        
        # Оставляем только первые "rank" сингулярных значений
        U_approx = U[:, :rank]
        S_approx = torch.diag(S[:rank])
        Vh_approx = Vh[:rank, :]
        
        # Формируем аппроксимированную матрицу
        approx_matrix = torch.matmul(U_approx, torch.matmul(S_approx, Vh_approx))
        approx_matrices.append(approx_matrix)

    return torch.stack(approx_matrices)

Q_approx = svd_approximation(Q, rank)
K_approx = svd_approximation(K, rank)
V_approx = svd_approximation(V, rank)

# Вычисление внимания
def compute_attention(Q, K, V, d_k):
    # Скаларное произведение запросов и ключей
    attention_scores = torch.matmul(Q, K.transpose(-2, -1)) / np.sqrt(d_k)
    
    # Применяем softmax для нормализации внимания
    attention_weights = F.softmax(attention_scores, dim=-1)
    
    # Вычисляем итоговое значение внимания
    output = torch.matmul(attention_weights, V)
    return output, attention_weights

# Многоголовочное внимание
def multi_head_attention(Q, K, V, num_heads, d_k):
    # Разделяем запросы, ключи и значения на несколько голов
    batch_size, seq_len, embed_dim = Q.shape
    head_dim = embed_dim // num_heads  # Размерность каждой головы
    
    # Изменяем форму данных для разделения на головы
    Q_ = Q.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    K_ = K.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    V_ = V.view(batch_size, seq_len, num_heads, head_dim).transpose(1, 2)
    
    # Вычисляем внимание для каждой головы
    attention_outputs = []
    attention_weights_list = []
    
    for i in range(num_heads):
        output, attention_weights = compute_attention(Q_[:, i], K_[:, i], V_[:, i], d_k)
        attention_outputs.append(output)
        attention_weights_list.append(attention_weights)
    
    # Объединяем результаты всех голов
    attention_output = torch.cat(attention_outputs, dim=-1)  # Сшиваем по размерности "embed_dim"
    
    # Линейное преобразование (процесс после многоголовочного внимания)
    output = attention_output.view(batch_size, seq_len, embed_dim)
    
    return output, attention_weights_list

# Размерность ключей
d_k = Q.shape[-1]

# Вычисление многоголовочного внимания
output, attention_weights_list = multi_head_attention(Q_approx, K_approx, V_approx, num_heads, d_k)

# Добавление позиционных кодов
output_with_pos = output + positional_encoding.unsqueeze(0)  # Добавляем позиционные коды к выходу

# Применение нормализации
def layer_normalization(x, eps=1e-6):
    mean = x.mean(dim=-1, keepdim=True)
    std = x.std(dim=-1, keepdim=True)
    return (x - mean) / (std + eps)

# Применяем нормализацию
normalized_output = layer_normalization(output_with_pos)

# Пропускное соединение (skip connection)
final_output = normalized_output + output  # Добавляем исходный выход (residual connection)

# Визуализация карты внимания для первой головы
plt.figure(figsize=(6, 6))
plt.imshow(attention_weights_list[0][0].detach().cpu().numpy(), cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Attention Map for Head 1 (SVD Approximation)')
plt.xlabel('Keys')
plt.ylabel('Queries')
plt.show()

# Вывод итогового результата
print("Output shape after multi-head attention, positional encoding, and normalization:", final_output.shape)
```

```python
# Полносвязный слой (Feed-Forward Network)
def feed_forward_network(embed_dim, hidden_dim):
    return torch.nn.Sequential(
        torch.nn.Linear(embed_dim, hidden_dim),  # Первый линейный слой
        torch.nn.ReLU(),  # Активация
        torch.nn.Linear(hidden_dim, embed_dim)  # Второй линейный слой
    )

# Параметры для FFN
hidden_dim = 512  # Размер скрытого слоя

# Инициализация FFN
ffn = feed_forward_network(embed_dim, hidden_dim)

# Применяем FFN
ffn_output = ffn(final_output)

# Применение нормализации к выходу FFN
normalized_ffn_output = layer_normalization(ffn_output)

# Пропускное соединение после FFN
final_ffn_output = normalized_ffn_output + final_output

# Логика для дальнейшего применения, например, классификации или другого вида задачи
# Здесь мы просто выводим размерность результата
print("Final output shape after Feed Forward and residual connection:", final_ffn_output.shape)

# Если нужна классификация, можно применить выходной слой:
# output_layer = torch.nn.Linear(embed_dim, num_classes)
# classification_output = output_layer(final_ffn_output)
```

```python
class MultiHeadAttention(torch.nn.Module):
    def __init__(self, embed_dim, num_heads):
        super(MultiHeadAttention, self).__init__()
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads

        # Проверка, что размерность эмбеддинга делится на количество голов
        assert embed_dim % num_heads == 0, "Размерность эмбеддинга должна делиться на количество голов"

        # Слой для получения Q, K, V
        self.query = torch.nn.Linear(embed_dim, embed_dim)
        self.key = torch.nn.Linear(embed_dim, embed_dim)
        self.value = torch.nn.Linear(embed_dim, embed_dim)

        # Финальная линейная трансформация после внимания
        self.fc_out = torch.nn.Linear(embed_dim, embed_dim)

    def forward(self, query, key, value):
        batch_size = query.shape[0]

        # Проектирование Q, K, V
        Q = self.query(query)
        K = self.key(key)
        V = self.value(value)

        # Разбиение на несколько голов
        Q = Q.view(batch_size, -1, self.num_heads, self.head_dim).transpose(1, 2)
        K = K.view(batch_size, -1, self.num_heads, self.head_dim).transpose(1, 2)
        V = V.view(batch_size, -1, self.num_heads, self.head_dim).transpose(1, 2)

        # Масштабированное произведение скалярного произведения
        energy = torch.matmul(Q, K.transpose(-2, -1))  # (batch_size, num_heads, seq_len, seq_len)
        attention = torch.nn.functional.softmax(energy / (self.head_dim ** 0.5), dim=-1)

        # Применение внимания
        out = torch.matmul(attention, V)  # (batch_size, num_heads, seq_len, head_dim)

        # Объединение голов
        out = out.transpose(1, 2).contiguous().view(batch_size, -1, self.num_heads * self.head_dim)

        # Применение финальной линейной трансформации
        out = self.fc_out(out)
        
        return out
```

```python
# Глобальная нормализация
class TransformerBlock(torch.nn.Module):
    def __init__(self, embed_dim, hidden_dim, num_heads):
        super(TransformerBlock, self).__init__()
        self.attention = MultiHeadAttention(embed_dim, num_heads)
        self.ffn = feed_forward_network(embed_dim, hidden_dim)
        self.layer_norm1 = torch.nn.LayerNorm(embed_dim)
        self.layer_norm2 = torch.nn.LayerNorm(embed_dim)
        self.dropout = torch.nn.Dropout(0.1)  # Dropout для регуляризации

    def forward(self, x):
        # Применение многоголовочного внимания с нормализацией
        attention_output = self.attention(x, x, x)
        attention_output = self.dropout(attention_output)
        attention_output = self.layer_norm1(attention_output + x)

        # Применение FFN с нормализацией
        ffn_output = self.ffn(attention_output)
        ffn_output = self.dropout(ffn_output)
        ffn_output = self.layer_norm2(ffn_output + attention_output)

        return ffn_output

# Параметры трансформер блока
num_heads = 8  # Число голов для многоголовочного внимания
embed_dim = 128  # Размерность эмбеддинга
hidden_dim = 512  # Размер скрытого слоя в FFN

# Инициализация трансформер блока
transformer_block = TransformerBlock(embed_dim, hidden_dim, num_heads)

# Пример входных данных
input_data = torch.randn(batch_size, seq_len, embed_dim)

# Прогон данных через блок
output_data = transformer_block(input_data)

print("Final output shape after transformer block:", output_data.shape)
```

```python
import torch
import torch.nn as nn

# Определение функции для получения позиционного кодирования
def get_positional_encoding(seq_len, embed_dim):
    positions = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)  # Размерность: [seq_len, 1]
    div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * -(torch.log(torch.tensor(10000.0)) / embed_dim))  # Размерность: [embed_dim // 2]
    
    # Позиционные коды
    angles = positions * div_term  # Размерность: [seq_len, embed_dim // 2]
    
    positional_encoding = torch.zeros(seq_len, embed_dim)  # Размерность: [seq_len, embed_dim]
    positional_encoding[:, 0::2] = torch.sin(angles)  # Четные индексы
    positional_encoding[:, 1::2] = torch.cos(angles)  # Нечетные индексы
    
    return positional_encoding
```

```python
class TransformerWithPositionalEncoding(torch.nn.Module):
    def __init__(self, embed_dim, hidden_dim, num_heads, seq_len):
        super(TransformerWithPositionalEncoding, self).__init__()
        self.positional_encoding = get_positional_encoding(seq_len, embed_dim)
        self.transformer_block = TransformerBlock(embed_dim, hidden_dim, num_heads)

    def forward(self, x):
        # Добавление позиционного кодирования к входным данным
        x = x + self.positional_encoding.to(x.device)
        
        # Прогон через блок трансформера
        return self.transformer_block(x)

# Параметры модели
seq_len = 128  # Длина последовательности
embed_dim = 128  # Размерность эмбеддинга
hidden_dim = 512  # Размер скрытого слоя в FFN
num_heads = 8  # Число голов для многоголовочного внимания

# Инициализация модели с позиционными кодами
model = TransformerWithPositionalEncoding(embed_dim, hidden_dim, num_heads, seq_len)

# Пример входных данных
input_data = torch.randn(batch_size, seq_len, embed_dim)

# Прогон данных через модель
output_data = model(input_data)

print("Final output shape after transformer with positional encoding:", output_data.shape)
```

```python
class FeedForwardNetwork(torch.nn.Module):
    def __init__(self, embed_dim, hidden_dim):
        super(FeedForwardNetwork, self).__init__()
        self.fc1 = torch.nn.Linear(embed_dim, hidden_dim)
        self.relu = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(hidden_dim, embed_dim)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
class TransformerBlockWithFFN(torch.nn.Module):
    def __init__(self, embed_dim, hidden_dim, num_heads):
        super(TransformerBlockWithFFN, self).__init__()
        self.attention = MultiHeadAttention(embed_dim, num_heads)
        self.ffn = FeedForwardNetwork(embed_dim, hidden_dim)
        self.layer_norm1 = torch.nn.LayerNorm(embed_dim)
        self.layer_norm2 = torch.nn.LayerNorm(embed_dim)

    def forward(self, x):
        # Многоголовочное внимание
        attention_output = self.attention(x, x, x)
        # Остаточное соединение и нормализация
        x = self.layer_norm1(x + attention_output)
        
        # Применение сети обратной связи (FFN)
        ffn_output = self.ffn(x)
        # Остаточное соединение и нормализация
        x = self.layer_norm2(x + ffn_output)
        
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import StepLR
from tqdm import tqdm  # Импортируем tqdm для отображения прогресса

class TransformerWithFullComponents(nn.Module):
    def __init__(self, embed_dim, num_heads, num_layers, num_classes):
        super(TransformerWithFullComponents, self).__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.num_layers = num_layers
        self.num_classes = num_classes

        # Используем слой для преобразования изображения в последовательности
        self.conv = nn.Conv2d(in_channels=3, out_channels=embed_dim, kernel_size=3, stride=1, padding=1)

        # Определение трансформера
        self.transformer_encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=embed_dim, nhead=num_heads),
            num_layers=num_layers
        )
        self.fc = nn.Linear(embed_dim, num_classes)

    def forward(self, x):
        # Проходим через сверточный слой для получения признаков
        x = self.conv(x)  # x: [batch_size, embed_dim, height, width]

        # Преобразуем тензор в форму [seq_len, batch_size, embed_dim]
        x = x.flatten(2).permute(2, 0, 1)  # seq_len = height * width

        # Прогоняем через трансформер
        x = self.transformer_encoder(x)

        # Применяем линейный слой
        x = self.fc(x[-1, :, :])  # Применяем к последнему элементу последовательности

        return x


# Гиперпараметры
embed_dim = 512
num_heads = 8
num_layers = 6
num_classes = 10  # Для примера

# Модель, оптимизатор и scheduler
model = TransformerWithFullComponents(embed_dim=512, num_heads=8, num_layers=6, num_classes=10)
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = StepLR(optimizer, step_size=2, gamma=0.7)

# Пример данных (64 изображения, размер 224x224, 3 канала)
inputs = torch.randn(64, 3, 224, 224)  # Пример случайных данных
labels = torch.randint(0, num_classes, (64,))  # Пример случайных меток

criterion = nn.CrossEntropyLoss()

# Тренировочный цикл с прогрессом
for epoch in tqdm(range(5), desc="Epochs", ncols=100):  # tqdm для отслеживания прогресса по эпохам
    model.train()
    optimizer.zero_grad()
    
    # Преобразование типа данных для inputs, если они не в LongTensor
    inputs = inputs.float()  # Преобразуем к float32 (обычно для изображений)
    
    outputs = model(inputs)  # Прогон через модель
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    scheduler.step()  # Обновление learning rate
    tqdm.write(f"Epoch [{epoch+1}/5], Loss: {loss.item():.4f}")  # Печать информации о текущем состоянии
```

```python
"тестинг"
import torch
import torch.nn as nn
from torchvision.models import vit_b_16
from torch import Tensor

# Класс для ViT с низкоранговыми аппроксимациями
class NLAEnhancedViT(nn.Module):
    def __init__(self, vit_model: nn.Module, rank: int = 128):
        super(NLAEnhancedViT, self).__init__()

        # Загружаем ViT без финального слоя
        self.conv_proj = vit_model.conv_proj
        self.encoder = vit_model.encoder

        # Позиционные эмбеддинги
        self.pos_embedding = vit_model.encoder.pos_embedding
        self.cls_token = vit_model.class_token  # [CLS] токен
        self.dropout = vit_model.encoder.dropout

        # Число входных признаков перед финальным слоем
        self.num_features = vit_model.heads.head.in_features
        print(f"Number of features from ViT: {self.num_features}")

        # Матрица для низкоранговой аппроксимации
        self.low_rank_matrix = nn.Parameter(torch.randn(self.num_features, rank))

        # Линейная трансформация для вывода
        self.fc = nn.Linear(rank, vit_model.heads.head.out_features)

    def forward(self, x: Tensor) -> Tensor:
        # Преобразуем изображение в последовательность патчей
        x = self.conv_proj(x)  # [batch_size, hidden_dim, height, width]
        x = x.flatten(2).transpose(1, 2)  # [batch_size, seq_length, hidden_dim]

        # Добавляем [CLS] токен
        batch_size = x.size(0)
        cls_tokens = self.cls_token.expand(batch_size, -1, -1)  # [batch_size, 1, hidden_dim]
        x = torch.cat((cls_tokens, x), dim=1)  # [batch_size, seq_length + 1, hidden_dim]

        # Добавляем позиционные эмбеддинги
        x = x + self.pos_embedding
        x = self.dropout(x)

        # Прогоняем через энкодер
        x = self.encoder(x)

        # Усредняем признаки по патчам
        x = x[:, 0]  # Используем [CLS] токен для представления изображения

        # Применяем низкоранговую аппроксимацию
        x = torch.matmul(x, self.low_rank_matrix)

        # Финальная линейная трансформация
        x = self.fc(x)

        return x


# Загружаем предобученную ViT модель
vit_model = vit_b_16(weights="IMAGENET1K_V1")  # Используем обновленный параметр weights

# Инициализация улучшенной модели
model_nla = NLAEnhancedViT(vit_model)

# Перевод модели на GPU, если доступно
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_nla = model_nla.to(device)

# Пример подачи изображения
input_image = torch.randn(1, 3, 224, 224).to(device)
output = model_nla(input_image)

# Выводим размерность выходных данных
print("Output shape:", output.shape)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CholeskyEnhancedSelfAttention(nn.Module):
    def __init__(self, embed_dim, num_heads):
        super(CholeskyEnhancedSelfAttention, self).__init__()
        self.num_heads = num_heads
        self.embed_dim = embed_dim

        self.q_linear = nn.Linear(embed_dim, embed_dim)
        self.k_linear = nn.Linear(embed_dim, embed_dim)
        self.v_linear = nn.Linear(embed_dim, embed_dim)
        self.out_linear = nn.Linear(embed_dim, embed_dim)

    def cholesky_decomposition(self, matrix):
        """
        Apply Cholesky decomposition to a matrix (if positive definite).
        """
        try:
            L = torch.cholesky(matrix)
            return L
        except RuntimeError:
            # If the matrix is not positive definite, return the original matrix (fallback).
            return matrix

    def forward(self, query, key, value):
        # Apply linear transformations
        Q = self.q_linear(query)
        K = self.k_linear(key)
        V = self.v_linear(value)

        # Reshape Q, K, V for multi-head attention
        batch_size, seq_len, _ = Q.shape
        Q = Q.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)
        K = K.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)
        V = V.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)

        # Compute attention scores
        attn_scores = torch.matmul(Q, K.transpose(-2, -1)) / (self.embed_dim // self.num_heads) ** 0.5

        # Apply Cholesky decomposition to attention logits
        L = self.cholesky_decomposition(attn_scores)
        attn_probs = F.softmax(L, dim=-1)  # Softmax over the last dimension

        # Compute the output by applying attention probabilities
        attn_output = torch.matmul(attn_probs, V)

        # Reshape the output back
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, self.embed_dim)

        # Final linear transformation
        output = self.out_linear(attn_output)

        return output

# Example of replacing self-attention in a Vision Transformer model
for name, layer in vit_model.encoder.layers.named_children():
    if hasattr(layer, 'self_attention'):
        # Replace the self-attention with CholeskyEnhancedSelfAttention
        layer.self_attention = CholeskyEnhancedSelfAttention(
            embed_dim=layer.self_attention.out_proj.in_features, 
            num_heads=layer.self_attention.num_heads
        )
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
import time
import torchvision
import torchvision.transforms as transforms
from torchvision import datasets, models

# Define the CholeskyEnhancedSelfAttention class
class CholeskyEnhancedSelfAttention(nn.Module):
    def __init__(self, embed_dim, num_heads):
        super(CholeskyEnhancedSelfAttention, self).__init__()
        self.num_heads = num_heads
        self.embed_dim = embed_dim

        self.q_linear = nn.Linear(embed_dim, embed_dim)
        self.k_linear = nn.Linear(embed_dim, embed_dim)
        self.v_linear = nn.Linear(embed_dim, embed_dim)
        self.out_linear = nn.Linear(embed_dim, embed_dim)

    def cholesky_decomposition(self, matrix):
        try:
            L = torch.cholesky(matrix)
            return L
        except RuntimeError:
            return matrix

    def forward(self, query, key, value):
        Q = self.q_linear(query)
        K = self.k_linear(key)
        V = self.v_linear(value)

        batch_size, seq_len, _ = Q.shape
        Q = Q.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)
        K = K.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)
        V = V.view(batch_size, seq_len, self.num_heads, self.embed_dim // self.num_heads).transpose(1, 2)

        attn_scores = torch.matmul(Q, K.transpose(-2, -1)) / (self.embed_dim // self.num_heads) ** 0.5

        L = self.cholesky_decomposition(attn_scores)
        attn_probs = F.softmax(L, dim=-1)

        attn_output = torch.matmul(attn_probs, V)
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, self.embed_dim)
        output = self.out_linear(attn_output)
        return output

# Load CIFAR-10 dataset
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to 224x224
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)


# Load a pre-trained ViT model and replace self-attention layer
vit_model = models.vision_transformer.vit_b_16(pretrained=True)
for name, layer in vit_model.encoder.layers.named_children():
    if hasattr(layer, 'self_attention'):
        layer.self_attention = CholeskyEnhancedSelfAttention(embed_dim=layer.self_attention.out_proj.in_features,
                                                              num_heads=layer.self_attention.num_heads)

# Training function
def train_model(model, dataloader, optimizer, criterion):
    model.train()
    running_loss = 0.0
    for inputs, labels in dataloader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    return running_loss / len(dataloader)

# Inference function
def evaluate_model(model, dataloader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return correct / total

# Baseline training with ViT (without Cholesky enhancement)
def run_baseline():
    model = models.vision_transformer.vit_b_16(pretrained=True)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    start_time = time.time()
    for epoch in range(1):
        loss = train_model(model, train_loader, optimizer, criterion)
    baseline_training_time = time.time() - start_time
    print(f"Baseline training time: {baseline_training_time} seconds")

    start_time = time.time()
    accuracy = evaluate_model(model, test_loader)
    baseline_inference_time = time.time() - start_time
    print(f"Baseline inference time: {baseline_inference_time} seconds")
    print(f"Baseline accuracy: {accuracy * 100:.2f}%")

# Modified training with Cholesky-enhanced ViT
def run_modified():
    optimizer = optim.Adam(vit_model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    start_time = time.time()
    for epoch in range(1):
        loss = train_model(vit_model, train_loader, optimizer, criterion)
    modified_training_time = time.time() - start_time
    print(f"Modified training time: {modified_training_time} seconds")

    start_time = time.time()
    accuracy = evaluate_model(vit_model, test_loader)
    modified_inference_time = time.time() - start_time
    print(f"Modified inference time: {modified_inference_time} seconds")
    print(f"Modified accuracy: {accuracy * 100:.2f}%")

# Run the baseline and modified model
run_baseline()
run_modified()
```
